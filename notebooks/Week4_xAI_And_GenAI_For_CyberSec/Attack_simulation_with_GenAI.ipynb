```python
import openai
import os

# --- Configuration ---
# It's highly recommended to set your API key as an environment variable.
# Example: export OPENAI_API_KEY='your-api-key'
# If you must hardcode for a quick test (NOT recommended for production):
# openai.api_key = "YOUR_ACTUAL_API_KEY"

try:
    openai.api_key = os.environ["OPENAI_API_KEY"]
    print("OpenAI API key loaded successfully.")
except KeyError:
    print("Error: OPENAI_API_KEY environment variable not set.")
    print("Please set your OpenAI API key before running this script.")
    # Exit or handle the error appropriately if the key is not found
    exit()

# --- LLM Model Selection ---
# We'll use a recent, capable model. Check OpenAI documentation for the latest.
# Common choices: 'gpt-4o', 'gpt-4-turbo', 'gpt-3.5-turbo'
LLM_MODEL = "gpt-4o"

print(f"Using LLM model: {LLM_MODEL}")

# --- Helper Function for API Calls ---
def generate_text(prompt, max_tokens=500, temperature=0.7):
    """Sends a prompt to the OpenAI API and returns the generated text."""
    try:
        response = openai.chat.completions.create(
            model=LLM_MODEL,
            messages=[
                {"role": "system", "content": "You are a helpful cybersecurity assistant specializing in generating realistic attack simulations."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=max_tokens,
            temperature=temperature, # Controls randomness: lower is more focused, higher is more creative
            n=1, # Number of completions to generate
            stop=None # Sequence where the API will stop generating further tokens
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"An error occurred during API call: {e}")
        return None

print("Environment setup complete. Ready to generate content.")
print("If you see this message → your lab is working perfectly!")
```

```python
# Crafting Prompts for a Sophisticated Phishing Email after above set up completion
```

```python
# Assuming the setup from the previous cell is already run and openai is configured.

phishing_prompt = """
Objective: Generate a highly convincing spear-phishing email designed for a Business Email Compromise (BEC) attack simulation. The email should target an employee in the finance department of a mid-sized technology company.

Sender Persona: The email should appear to be from the Chief Financial Officer (CFO) of the company, 'Mr. Alan Sterling'.

Recipient Persona: The email is addressed to a member of the finance team, let's assume their name is 'Sarah Chen'.

Scenario: Mr. Sterling needs an urgent wire transfer to finalize a confidential acquisition deal that is time-sensitive and cannot be discussed via phone or in person due to its sensitive nature. He needs the transfer to be processed immediately today.

Key Information to Include:
- Sender: Alan Sterling, CFO
- Recipient: Sarah Chen
- Subject Line: Needs to be urgent and attention-grabbing, e.g., \"URGENT: Confidential Acquisition Wire Transfer\"
- Wire Transfer Amount: $75,000 USD
- Beneficiary Bank: Placeholder for a bank name (e.g., \"Global Trust Bank\")
- Account Number: Placeholder for an account number (e.g., \"GTB-XXXXXX1234\")
- Routing Number: Placeholder (e.g., \"071000013\")
- Purpose of Transfer: \"Confidential Acquisition - Project Nightingale\"
- Call to Action: Request Sarah to process the transfer immediately and confirm once done. Emphasize discretion.

Tone: Authoritative, urgent, and slightly stressed, reflecting the pressure of a time-sensitive deal. It should be professional but convey a sense of immediate need.

Constraints:
- Avoid any obvious grammatical errors or typos.
- Do not include any suspicious links or attachments. The request should be purely informational, prompting the recipient to initiate a transfer via standard company procedures (which the attacker hopes will be bypassed due to urgency).
- Ensure the language is professional and aligns with executive communication.

Generate the full email content, including a suitable subject line.
"""

print("Generating sophisticated phishing email...")

# Generate the email content
generated_email = generate_text(phishing_prompt, max_tokens=700, temperature=0.8)

if generated_email:
    print("\n--- Generated Phishing Email ---")
    print(generated_email)
    print("-------------------------------")
else:
    print("Failed to generate email.")

print("\nPhishing email generation complete.")
```
